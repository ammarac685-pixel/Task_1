```python
import pandas as pd

# Sample raw product data
data = [
    {"Product": "Laptop", "Category": "Electronics", "Price": "50000"},
    {"Product": "Mobile", "Category": "Electronics", "Price": "20000"},
    {"Product": "Laptop", "Category": "Electronics", "Price": "50000"},
    {"Product": "Chair", "Category": "Furniture", "Price": "5000"},
    {"Product": "Table", "Category": "Furniture", "Price": "10000"},
    {"Product": "Mobile", "Category": "Electronics", "Price": "20000"},
    {"Product": "TV", "Category": "Electronics", "Price": "40000"},
    {"Product": "Sofa", "Category": "Furniture", "Price": "15000"},
    {"Product": None, "Category": "Electronics", "Price": "30000"}
]

# Create a DataFrame from the raw data
df = pd.DataFrame(data)

# Clean the data
df = df.drop_duplicates()  # Remove duplicates
df = df.dropna(subset=["Product"])  # Remove rows with missing product names
df["Price"] = df["Price"].astype(float)  # Convert prices to float

# Group products by category
products_by_category = df.groupby("Category")

# Print a neat summary,
print("Product Catalogue Summary:")
for category, group in products_by_category:
    print(f"\n{category}:")
    print(group[["Product", "Price"]])
    print(f"Total Products: {len(group)}")
    print(f"Total Value: {group['Price'].sum():.2f}")
    print(f"Average Price: {group['Price'].mean():.2f}")
```
