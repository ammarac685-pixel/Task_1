```python
# Raw orders
orders = [
    {"order_id": 1, "customer": "Ammara", "category": "electronics", "amount": 500},
    {"order_id": 2, "customer": "Saira", "category": "clothing", "amount": 200},
    {"order_id": 3, "customer": "Nida", "category": "electronics", "amount": 90},
    {"order_id": 4, "customer": "Maria", "category": "groceries", "amount": 150},
    {"order_id": 5, "customer": "Zara", "category": "clothing", "amount": 80}
]

# Discount rules
discount_rules = {
    "electronics": 0.10,
    "clothing": 0.20,
    "groceries": 0.05
}

MIN_ORDER = 100


# Apply discount
discounted_orders = list(
    map(lambda o: {**o,
                   "discounted_amount": o["amount"] * (1 - discount_rules.get(o["category"], 0))},
        orders)
)

# Filter orders
filtered_orders = list(
    filter(lambda o: o["discounted_amount"] >= MIN_ORDER, discounted_orders)
)

# Total revenue
total_revenue = sum(map(lambda o: o["discounted_amount"], filtered_orders))


# -------- Pretty Output --------

print("="*55)
print("        ORDER DISCOUNT PIPELINE SUMMARY")
print("="*55)

print(f"{'ID':<5}{'Customer':<12}{'Category':<15}{'Original':<10}{'Final'}")
print("-"*55)

list(
    map(lambda o: print(f"{o['order_id']:<5}{o['customer']:<12}{o['category']:<15}{o['amount']:<10}{round(o['discounted_amount'],2)}"),
        filtered_orders)
)

print("-"*55)
print(f"TOTAL REVENUE: {round(total_revenue,2)}")
print("="*55)
```
