```python
products = [
    {"name": "Laptop", "category": "Electronics", "price": 1000},
    {"name": "Phone", "category": "Electronics", "price": 500},
    {"name": "Shirt", "category": "Fashion", "price": 20},
    {"name": "Laptop", "category": "Electronics", "price": 1000},  
    {"name": "Shoes", "category": "Fashion", "price": 50},
    {"name": "TV", "category": "Electronics", "price": 1500},
    {"name": "Watch", "category": "Fashion", "price": 200},
    {"name": "Headphones", "category": "Electronics", "price": 300},
    {"name": "Sunglasses", "category": "Fashion", "price": 100},
    {"name": "Book", "category": "Books", "price": 10},
    {"name": "Pen", "category": "Stationery", "price": 1},
    {"name": "Notebook", "category": "Stationery", "price": 5},
    {"name": "Bag", "category": "Fashion", "price": 80},
    {"name": "Camera", "category": "Electronics", "price": 800},
]

# Clean data and group by category
summary = {
    category: sorted(
        [p for p in products if p.get('category') == category and 'price' in p],
        key=lambda x: x['price'], reverse=True
    )[:5]
    for category in set(p['category'] for p in products if 'price' in p)
}

# Print with formatting
print("{")
for category, items in summary.items():
    print(f"  '{category}': [")
    for item in items:
        print(f"    {{'name': '{item['name']}', 'category': '{item['category']}', 'price': {item['price']}}},")
    print("  ],")
print("}")
```
