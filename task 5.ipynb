```python
# Sample data
orders = [
    {"customer_id": 1, "category": "Electronics", "amount": 500},
    {"customer_id": 2, "category": "Fashion", "amount": 200},
    {"customer_id": 3, "category": "Electronics", "amount": 1000},
    {"customer_id": 4, "category": "Fashion", "amount": 50},
    {"customer_id": 5, "category": "Electronics", "amount": 800},
]

# Discount rules
discount_rules = {
    "Electronics": 0.1,  # 10% discount
    "Fashion": 0.2,  # 20% discount
}

# Minimum threshold
min_threshold = 100

# Process orders
processed_orders = list(filter(
    lambda x: x["amount"] >= min_threshold,
    map(
        lambda x: {
            **x,
            "discount": x["amount"] * discount_rules.get(x["category"], 0),
            "revenue": x["amount"] * (1 - discount_rules.get(x["category"], 0)),
        },
        orders,
    ),
))

# Summary
total_revenue = sum(map(lambda x: x["revenue"], processed_orders))

print("Processed Orders:")
print("-" * 50)
print("{:<15} {:<15} {:<10} {:<10} {:<10}".format("Customer ID", "Category", "Amount", "Discount", "Revenue"))
print("-" * 50)
list(map(
    lambda x: print("{:<15} {:<15} {:<10} {:<10.2f} {:<10.2f}".format(
        x["customer_id"], x["category"], x["amount"], x["discount"], x["revenue"]
    )),
    processed_orders
))
print("-" * 50)
print(f"\nTotal Revenue: {total_revenue:.2f}")
```
