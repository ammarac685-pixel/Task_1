```python
import bisect

class Employee:
    def __init__(self, name, performance_score, experience):
        self.name = name
        self.performance_score = performance_score
        self.experience = experience

    def __repr__(self):
        return f"{self.name} (Score: {self.performance_score}, Exp: {self.experience})"

# Sample data
employees = [
    Employee("John", 90, 5),
    Employee("Alice", 90, 3),
    Employee("Bob", 80, 7),
    Employee("Charlie", 90, 4),
    Employee("David", 70, 10),
]

# Sort employees by performance score, then experience, then name
sorted_employees = sorted(employees, key=lambda x: (-x.performance_score, -x.experience, x.name))

# Extract names for binary search
names = [emp.name for emp in sorted_employees]

# Search for an employee by name
def search_employee(name):
    idx = bisect.bisect_left(names, name)
    if idx != len(names) and names[idx] == name:
        return sorted_employees[idx]
    return None

# Example usage
print("Ranked Employees:")
for emp in sorted_employees:
    print(emp)

# Search for an employee
name_to_search = "Alice"
result = search_employee(name_to_search)
print(f"\nSearch result for '{name_to_search}': {result}")
```
