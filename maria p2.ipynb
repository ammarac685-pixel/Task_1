```python
import pandas as pd
import os

data_folder = r"C:\Users\Maria\Desktop\student_result_project\data"
output_file = r"C:\Users\Maria\Desktop\student_result_project\output\clean_results.csv"

all_students = []

files = os.listdir(data_folder)

for file in files:
    try:
        path = os.path.join(data_folder, file)
        df = pd.read_csv(path)

        for index, row in df.iterrows():
            try:
                student_id = row['id']
                name = row['Name']   # Capital N
                marks = int(row['Marks'])   # Capital M

                status = "Pass"
                if marks < 40:
                    status = "Fail"

                all_students.append([student_id, name, marks, status])
            except:
                print("Invalid row skipped in", file)
    except:
        print("File could not be read:", file)

clean_df = pd.DataFrame(all_students, columns=["ID", "Name", "Marks", "Status"])
clean_df.to_csv(output_file, index=False)

print("Clean result file created!")
print(clean_df)
```
