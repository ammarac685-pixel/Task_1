```python
import pandas as pd
import datetime
import os

# Dummy data
audit_data = [
    ["file1.py", 10, "Python Script", "2022-01-01 12:00:00", "OK"],
    ["file2.txt", 5, "Text File", "2022-01-02 13:00:00", "Warning"],
    ["file3.csv", 20, "CSV File", "2022-01-03 14:00:00", "Error"],
    ["file4.ipynb", 15, "Jupyter Notebook", "2024-09-01 15:00:00", "Over 30 Days"],
]
penalties = [
    ["file4.ipynb", 15, "Overage Fine"],
]

# Reports folder
reports_folder = "reports"
os.makedirs(reports_folder, exist_ok=True)

timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")

audit_df = pd.DataFrame(audit_data, columns=["File Name", "Size (KB)", "Type", "Last Modified", "Flag"])
audit_df.to_csv(os.path.join(reports_folder, f"audit_report_{timestamp}.csv"), index=False)

if penalties:
    penalty_df = pd.DataFrame(penalties, columns=["File Name", "Size (KB)", "Overage Fine"])
    penalty_df.to_csv(os.path.join(reports_folder, f"penalties_report_{timestamp}.csv"), index=False)

print("Reports generated successfully!")
```
